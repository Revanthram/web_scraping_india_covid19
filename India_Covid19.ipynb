```python
'''
importing all the required libraries.
'''
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
'''
get complete html content of the website.
'''
source=requests.get("https://www.mohfw.gov.in/").text
soup=BeautifulSoup(source,"html.parser")
```

```python
'''
scrapeing the required tags and writing into the csv file.
'''
table=soup.find("table",class_="table table-striped")
states=[]
with open('India_Covid19.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["SL.No","State","Confirmed","Recovered","Deaths"])
    for row in table.find_all("tr"):
        for cell in row.find_all("td"):
            states.append(cell.text)
        writer.writerow(states)
        states=[]
        
```

```python
'''
reading the csv file to get data_frame through pandas.
'''
covid19=pd.read_csv("India_Covid19.csv")
covid19
```

```python
'''
droping the unwanted columns.
'''
covid19.dropna(how="any",inplace=True)
covid19
```

```python
'''
converting columns to the integer data type for visualization purpose.
'''
covid19["SL.No"]=covid19["SL.No"].astype("int64")
covid19["Confirmed"]=covid19["Confirmed"].astype("int64")
covid19["Recovered"]=covid19["Recovered"].astype("int64")
```

```python
'''
converting data cleaned data frame to a csv file
'''
pd.DataFrame(covid19).to_csv("India_covid19.csv")
```
